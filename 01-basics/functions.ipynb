```python
#Function that Doubles List Values
def double_all(data):
    result = []

    for i in data:
        doubled = i*2
        result.append(doubled)

    return result

print(double_all([2,3,4]))
```

```python
 #Function that Filters Passing Grades
def filter_passing(data):
    result = []
    
    for i in data:
        if i >= 70:
            result.append(i)

    return result

print(filter_passing([60, 75, 80, 50, 90]))
```

```python
#Function that Filters Values Greater Than Threshold

def filter_greater_than(data, threshold):
    result = []

    for x in data:
        if x > threshold:
            result.append(x)

    return result
    
data = [50, 120, 80, 200, 30]
print(filter_greater_than(data, 100))
```

```python
# Function that Counts Even Numbers

def count_even(numbers):
    count = 0

    for x in numbers:
        if x % 2 == 0:
            count += 1

    return count

numbers = [1,2,3,4,5,6]
print(count_even(numbers))
```

```python
#Function that Finds Maximum Value
def find_max(data):
    max_val = data[0]

    for x in data:
        if x > max_val:
            max_val = x

    return max_val

data = [10, 50, 30, 90, 20]
print(find_max(data))
```

```python
#Function that Checks if Value Exists in List
def check_value(data, target):
    for x in data:
        if x == target:
            return True
    return False

data = [1,3,5,7]
print(check_value(data, 5))
print(check_value(data, 10))
```

```python
#Function that Finds First Element Greater Than Threshold
def find_first_greater(data, threshold):
    for x in data:
        if x > threshold:
            return x 
    return None

data = [10, 20, 5, 30]
print(find_first_greater(data, 15))
```

####  PART 3 FUNCTIONS WITH MULTIPLE RETURNS

```python
#Function that Separates Positive and Negative Numbers
def separate_numbers(data):
    positive = []
    negative = []

    for x in data:
        if x > 0:
            positive.append(x)
        elif x < 0:
            negative.append(x)
    return positive, negative

data = [10, -5, 20, -3, 0]
print(separate_numbers(data))
```

```python
#Function with Conditional Return (Grade System)
def calculate_grade(score):
    if score >= 85:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 50:
        return'c'
    else:
        return "D"
    
print(calculate_grade(90))
print(calculate_grade(75))
print(calculate_grade(60))
print(calculate_grade(40))
```

```python
#Function that Returns Multiple Values (Discount System)
def check_discount(price):
    if price > 200000:
        discount_pct = 20
        final_price = price * 0.8
    elif price > 100000:
        discount_pct = 10
        final_price = price * 0.9
    else:
        discount_pct = 0
        final_price = price

    return price, discount_pct, final_price

print(check_discount(40000))
print(check_discount(150000))
print(check_discount(300000))
```

```python
#Function with Overtime Calculation

def total_salary(hours, hourly_rate):
    if hours <= 40:
        salary = hours * hourly_rate
        status = "normal"
    else:
        salary = (40 * hourly_rate) + ((hours - 40) * hourly_rate * 1.5)
        status = "overtime"

    return salary, status

print(total_salary(45, 10000))
print(total_salary(33, 10000))
```

```python
#Function with Employee Bonus Calculation

def calculate_bonus(salary, performance):
    if performance == "A":
        bonus = salary * 0.2
    elif performance == "B":
        bonus = salary * 0.1
    else:
        bonus = 0

    total = salary + bonus
    return total

print(calculate_bonus(5000000, "A"))
print(calculate_bonus(5000000, "B"))
print(calculate_bonus(5000000, "C"))
```

```python
#Function with Parking Fee Calculation
def calculate_parking(hours_parked):
    if hours_parked <= 2:
        fee = 5000
    else:
        fee = 5000 + (hours_parked - 2) * 2000
    return fee

print(calculate_parking(1))
print(calculate_parking(3))
print(calculate_parking(5))
```

#### PART 4 ADVANCED BUSINESS LOGIC FUNCTIONS

```python
#Function with Patient Priority System
def patient_priority(age, condition):
    if condition == "critical":
        priority = 1
    elif condition == "moderate":
        priority = 2
    else:
        priority = 3

    if age > 60:
        priority -= 1
        if priority < 1:
            priority = 1

    return priority

print(patient_priority(65, 'moderate'))
print(patient_priority(44, 'critical'))
print(patient_priority(70, 'mild'))
```

```python
#Function with Fraud Detection Logic
def check_fraud(amount, location):
    if amount > 100000 and location == 'foreign':
        return 'HIGH RISK'
    elif amount > 100000 or location == 'foreign':
        return 'SUSPICIOUS'
    else:
        return 'SAFE'
    
print(check_fraud(2000000, 'foreign'))
print(check_fraud(20000000, 'local'))
print(check_fraud(500000, 'foreign'))
print(check_fraud(500000, 'local'))
```

```python
#Function with Credit Approval Logic
def credit_approval(salary, loan, job_type):
    ratio = loan / salary

    if ratio <= 5 and job_type == 'permanent':
        status = 'APPROVED'
    elif ratio <= 5 and job_type == 'contract':
        status = 'REVIEW'
    else:
        status = 'REJECT'

    return salary, loan, ratio, status

print(credit_approval(1000000, 3000000, 'permanent'))
print(credit_approval(1000000, 4000000, 'contract'))
print(credit_approval(1000000, 7000000, 'permanent'))
print(credit_approval(1000000, 6000000, 'contract'))
```

```python
#Function with Loan Interest Calculation
def calculate_loan(loan_amount, status):
    if loan_amount <= 10000000:
        interest = 10
    else:
        interest = 15

    if status == "premium":
        interest -= 5

    total_payment = loan_amount + (loan_amount * interest / 100)

    return loan_amount, status, interest, total_payment

print(calculate_loan(3000000, "regular"))
print(calculate_loan(6000000, "premium"))
print(calculate_loan(12000000, "regular"))
print(calculate_loan(15000000, "premium"))
```

```python
#Function with Shipping Cost Calculator

weights = [1, 5, 12, 3, 20, 7, 15]

def calculate_shipping(weights):
    result = []
    count_heavy = 0

    for i in weights:
        if i > 10:
            cost = i * 5000
            result.append(cost)
            count_heavy += 1
        else:
            cost = i * 3000
            result.append(cost)

    return result, count_heavy

print(calculate_shipping(weights))
```

```python
#Function with Multi-List Shipping Categorization
weights = [1, 5, 12, 3, 20, 7, 15]

def categorize_shipping(weights):
    all_costs = []
    heavy = []
    light = []

    for i in weights:
        if i > 10:
            cost = i * 5000
            all_costs.append(cost)
            heavy.append(cost)
        else:
            cost = i * 3000
            all_costs.append(cost)
            light.append(cost)

    return all_costs, heavy, light

print(categorize_shipping(weights))
```

#### PART 5 ADVANCED DATA CLEANING & VALIDATION

```python
#Function with Try-Except for String Validation
def clean_numeric(data):
    clean = []

    for d in data:
        if d.isdigit():
            clean.append(int(d))

    return clean

data = ["10", "20", "error", "30"]
print(clean_numeric(data))
```

```python
#Function with Safe Division (Avoid zero division)
def safe_average(data):
    total = sum(data)
    count = len(data)

    if count > 0:
        return total / count
    else:
        return 0
    
print(safe_average([10,20,30]))
print(safe_average([]))
```

```python
#Function that Sums Only Valid Numbers
def sum_valid_numbers(data):
    total = 0

    for d in data:
        try:
            total += float(d)
        except:
            pass

    return total

data = ["10", "20", "error", "30"]

print(sum_valid_numbers(data))
```

```python
#Function that Counts Valid vs Invalid Entries
def count_valid_invalid(data):
    valid = 0
    invalid = 0

    for d in data:
        try:
            float(d)
            valid += 1
        except:
            invalid += 1

    return valid, invalid

data = ["10", "20", "abc", "30", "xyz"]

print(count_valid_invalid(data))
```

```python
#Function with Comprehensive Data Cleaning
def clean_data_stats(data):
    clean_data = []

    for d in data:
        try:
            clean_data.append(float(d))
        except:
            pass
    
    total = sum(clean_data)
    count_valid = len(clean_data)
    count_invalid = len(data) - count_valid
    average = total / count_valid if count_valid > 0 else 0

    return total, count_valid, count_invalid, average

data = ["100", "250.5", "300", "error", "150.75", "", "200"]

result = clean_data_stats(data)

print("Total:", result[0])
print("Valid:", result[1])
print("Invalid:", result[2])
print("Average:", result[3])
```

```python
#Function with Conditional String Combination
def combine_strings(data):
    result = ''

    for x in data:
        if x != 'b':
            result += x
        
    return result

data = ["a", "b", "c"]

print(combine_strings(data))
```
