## PART 2 — BASIC SET OPERATIONS

```python
# Remove duplicates from list using set
data = [1, 2, 2, 3, 4, 4, 5]

unique_data = set(data)

print("Original list:", data)
print("Unique values:", unique_data)
```

```python
# Find common elements between two sets
A = {1, 2, 3}
B = {2, 3, 4}

print("Set A:", A)
print("Set B:", B)
print("Intersection (A & B):", A & B)
```

```python
# Combine two sets
A = {1, 2, 3}
B = {2, 3, 4}

print("Set A:", A)
print("Set B:", B)
print("Union (A | B):", A | B)
```

```python
# Find difference between sets
A = {1, 2, 3}
B = {2, 3, 4}

print("Set A:", A)
print("Set B:", B)
print("A - B:", A - B)
print("B - A:", B - A)
```

```python
# Check if one set is subset/superset of another
A = {1, 2}
B = {1, 2, 3}

print("Set A:", A)
print("Set B:", B)
print("Is A subset of B?", A.issubset(B))
print("Is B superset of A?", B.issuperset(A))
```

```python
# Add new element to set
V = {'A', 'B'}
V.add('C')

print("Updated set:", V)
```

```python
# Check if element exists in set
auth_codes = {'A123', 'B456', 'C789'}

print("Is 'B456' authorized?", 'B456' in auth_codes)
print("Is 'X999' authorized?", 'X999' in auth_codes)
```

## PART 3 PRACTICAL APPLICATIONS

```python
# Find common interests between users
user_A = {"Python", "AI"}
user_B = {"Python", "Data"}

common_interests = user_A & user_B

print("User A interests:", user_A)
print("User B interests:", user_B)
print("Common interests:", common_interests)
```

```python
# Combine two lists and get unique values
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

set1 = set(list1)
set2 = set(list2)

unique_combined = set1 | set2

print("List 1:", list1)
print("List 2:", list2)
print("Combined unique values:", unique_combined)
```

```python
# Compare sum of list with duplicates vs set
A = [1, 2, 2, 1]
B = set(A)

print("List A:", A)
print("Set B:", B)
print("Sum of A:", sum(A))
print("Sum of B:", sum(B))
print("Are sums equal?", sum(A) == sum(B))
```

## PART 4 ADVANCED EXERCISES WITH DICTIONARIES

```python
# Filter customers by age and spending
customers = [
    {"age": 17, "spending": 2000},
    {"age": 25, "spending": 7000},
    {"age": 40, "spending": 15000},
    {"age": 16, "spending": 1000},
    {"age": 30, "spending": 8000},
    {"age": 22, "spending": 3000}
]

print("Customers with age >= 18 and spending >= 5000:")
for customer in customers:
    if customer['age'] >= 18 and customer['spending'] >= 5000:
        print(customer)
```

```python
# Evaluate employee attendance and categorize
data = [
    {"name": "Andi", "attendance": 22},
    {"name": "Budi", "attendance": 18},
    {"name": "Citra", "attendance": 25},
    {"name": "Dina", "attendance": 15}
]

def evaluate_attendance(data):
    result = []
    
    for d in data:
        if d["attendance"] >= 20:
            status = "Excellent"
        elif d["attendance"] >= 15:
            status = "Good"
        else:
            status = "Poor"
        
        result.append({"name": d["name"], "status": status})
    
    return result

print(evaluate_attendance(data))
```

```python
# Determine patient priority based on age and condition
def patient_priority(age, condition):
    if condition == "critical":
        priority = 1
    elif condition == "moderate":
        priority = 2
    else:
        priority = 3

    # Elderly patients get higher priority
    if age > 60:
        priority -= 1
        if priority < 1:
            priority = 1

    return priority

# Test cases
print("Priority (age 65, moderate):", patient_priority(65, 'moderate'))
print("Priority (age 44, critical):", patient_priority(44, 'critical'))
print("Priority (age 70, mild):", patient_priority(70, 'mild'))
```

```python
# Determine credit approval based on salary, loan amount, and job type
def credit_approval(salary, loan, job_type):
    ratio = loan / salary

    if ratio <= 5 and job_type == 'permanent':
        status = 'APPROVED'
    elif ratio <= 5 and job_type == 'contract':
        status = 'REVIEW'
    else:
        status = 'REJECT'

    return salary, loan, ratio, status

# Test cases
print(credit_approval(1000000, 3000000, 'permanent'))
print(credit_approval(1000000, 4000000, 'contract'))
print(credit_approval(1000000, 7000000, 'permanent'))
print(credit_approval(1000000, 6000000, 'contract'))
```

```python
# Calculate loan with interest based on amount and customer status
def calculate_loan(loan_amount, status):
    if loan_amount <= 10000000:
        interest = 10
    else:
        interest = 15

    if status == "premium":
        interest -= 5

    total_payment = loan_amount + (loan_amount * interest / 100)

    return loan_amount, status, interest, total_payment

# Test cases
print(calculate_loan(3000000, "regular"))
print(calculate_loan(6000000, "premium"))
print(calculate_loan(12000000, "regular"))
print(calculate_loan(15000000, "premium"))
```

## PART 5 DATA CLEANING AND VALIDATION

```python
# Clean data with errors and calculate statistics
data = ["100", "250.5", "300", "error", "150.75", "", "200"]

clean_data = []

for d in data:
    try:
        clean_data.append(float(d))
    except:
        pass

total = sum(clean_data)
count_valid = len(clean_data)
count_invalid = len(data) - count_valid
average = total / count_valid if count_valid > 0 else 0

print("Total:", total)
print("Valid entries:", count_valid)
print("Invalid entries:", count_invalid)
print("Average:", average)
```

```python
# Validate login based on username and password validity
username_valid = [True, True, False, True, False]
password_valid = [True, False, True, True, False]

print("Valid users:")
for i in range(len(username_valid)):
    if username_valid[i] == True and password_valid[i] == True:
        print(f'User index {i}: Valid')
```

```python
# Check discount eligibility based on membership or coupon
is_member = [True, False, True, False, True]
has_coupon = [False, True, False, False, True]

print("Users eligible for discount:")
for i, (m, c) in enumerate(zip(is_member, has_coupon)):
    if m or c:
        print(f'User {i} gets discount')
```

## PART 6 LOOP CONTROL AND DATA STREAMS

```python
# Find and display error log entries with their indices
logs = ["ok", "error", "error", "ok", "warning", "error", "ok"]

print("Error logs:")
for i in range(len(logs)):
    if logs[i] == 'error':
        print(f"Index {i}: {logs[i]}")
```

```python
# Process data stream until termination signal
data_stream = [5, 10, 15, 20, -999, 30, 40]

print("Processing data until stop signal (-999):")
for i in data_stream:
    if i == -999:
        print("Stop signal received")
        break
    print(i)
```

```python
# Process transactions until stop marker (-1)
transactions = [100, 200, 300, -1, 500, 600]

print("Processing transactions:")
i = 0
while transactions[i] != -1:
    print(f"Transaction {i}: {transactions[i]}")
    i += 1

print("Stop marker (-1) reached")
```
