## PART 1 CONDITIONS AND LOOPS - PRACTICE EXERCISES

```python
#Age Category (if-elif-else)
age = 18

if age <= 19:
    print("AC/DC")
elif age == 18:
    print("Pink Floyd")
else:
    print("Other")
```

```python
#Logical Operator (AND)
has_id = True
fingerprint = True

if has_id and fingerprint:
    print("Access granted")
else:
    print("Access denied")
```

 ## PART 2 BASIC LOOPS

```python
#For loop (basic iteration)
for num in range(1, 6):
    print(num)
```

```python
#While loop (o to 4)
i = 0

while i < 5:
    print(i)
    i += 1
```

```python
#While loop 
count = 10

while count > 0:
    print(count)
    count -= 1
```

## PART 3 LOOP + CONDITION

```python
#Filter even numbers
numbers = [1, 2, 3, 4, 5, 6]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

```python
#Sum of Even Numbers
numbers = [1, 2, 3, 4, 5, 6]

total = 0

for n in numbers:
    if n % 2 == 0:
        total += n

print(total)
```

```python
#Find Specific Value
data = [3, 7, 2, 9, 5]

for d in data:
    if d == 9:
        print("Found")
```

```python
#Check Value Exists
data = [1, 3, 5, 7]

found = False

for x in data:
    if x == 5:
        found = True

print(found)
```

```python
#Filtering Values > 100
data = [50, 120, 80, 200, 30]

result = []

for x in data:
    if x > 100:
        result.append(x)

print(result)
```
