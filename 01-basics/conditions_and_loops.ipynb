## PART 1 CONDITIONS AND LOOPS - PRACTICE EXERCISES

```python
#Age Category (if-elif-else)
age = 18

if age <= 19:
    print("AC/DC")
elif age == 18:
    print("Pink Floyd")
else:
    print("Other")
```

 ## PART 2 BASIC LOOPS

```python
#For loop (basic iteration)
for num in range(1, 6):
    print(num)
```

```python
#While loop (o to 4)
i = 0

while i < 5:
    print(i)
    i += 1
```

## PART 3 LOOP + CONDITION

```python
#Filter even numbers
numbers = [1, 2, 3, 4, 5, 6]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

```python
#Sum of Even Numbers
numbers = [1, 2, 3, 4, 5, 6]

total = 0

for n in numbers:
    if n % 2 == 0:
        total += n

print(total)
```

```python
#Find Specific Value
data = [3, 7, 2, 9, 5]

for d in data:
    if d == 9:
        print("Found")
```

```python
#Check Value Exists
data = [1, 3, 5, 7]

found = False

for x in data:
    if x == 5:
        found = True

print(found)
```

```python
#Filtering Values > 100
data = [50, 120, 80, 200, 30]

result = []

for x in data:
    if x > 100:
        result.append(x)

print(result)
```

## PART 4 LOOP CONTROL

```python
# Continue statement
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
```

```python
#Break Statement
for num in range(1, 10):
    if num == 5:
        break
    print(num)
```

## PART 5 INDEXING AND ENUMERATION

```python
#Index and value (enumerate)
squares = [1,4,9,16]

for i, square in enumerate(squares):
    print(i, square)
```

```python
#Index and value (manual)
logs = ["error", "ok", 'ok', 'error', 'waiting']

for i in range(len(logs)):
    print(i,(logs[i]))
```

```python
#filter error logs
logs = ["error", "ok", "ok", "error", "warning"]

for i in range(len(logs)):
    if logs[i] == "error":
        print(i, logs[i])
```

## PART 6 NESTED CONDITIONS

```python
#Nested conditions
numbers = [10, 15, 20, 25]

for n in numbers:
    if n % 2 == 0:
        if n > 15:
            print(n)
```

```python
#Positive and Negative Classification
data = [10, -5, 20, -3, 0]

positive = []
negative = []

for x in data:
    if x > 0:
        positive.append(x)
    elif x < 0:
        negative.append(x)

print("positive:", positive)
print("negative:", negative)
```

## PART 7 DATA PROCESSING

```python
#Multiply List Values
data = [1, 2, 3, 4]

result = []

for x in data:
    result.append(x * 2)

print(result)
```

```python
#Find Maximum Value
data = [10,50,40,90,20]

max_val = data[0]

for x in data:
    if x > max_val:
        max_val = x

print(max_val)
```

## PART 8 WHILE + DATA STREAM

```python
#Stop Condition (while)
transactions = [100, 200, 300, -1, 500, 600]

i = 0

while transactions[i] != -1:
    print(transactions[i])
    i += 1
```

```python
#stop Condition (for + break)

transactions = [100, 200, 300, -1, 500, 600]

for t in transactions:
    if t == -1:
        break
    print(t)
```

```python
#Transaction Limit Control
transactions = [50000, 120000, 3000000, 45000, 70000]

for t in transactions:
    if t > 100000:
        break
    print('Normal transaction:', t)
```

## PART 9 STRING & DATA FILTERING

```python
#Combine String with Condition
data = ["a", "b", "c"]

result = ""

for x in data:
    if x != "b":
        result += x

print(result)
```

```python
#Digit Validation
data = ["10", "20", "error", "30"]

clean = []

for d in data:
    if d.isdigit():
        clean.append(int(d))

print(clean)
```

## PART 10 ADVANCED DATA CLEANING

```python
#clean mixed data
data = ["100", "250.5", "300", "error", "150.75", "", "200"]

clean_data = []

for d in data:
    try:
        clean_data.append(float(d))
    except:
        pass

total = sum(clean_data)
count_valid = len(clean_data)
count_invalid = len(data) - count_valid
average = total / count_valid if count_valid > 0 else 0

print('total:', total)
print('valid:', count_valid)
print('invalid:', count_invalid)
print('average:', average)
```

```python
#Count Valid vs Invalid
data = ["10", "20", "abc", "30", "xyz"]

valid = 0
invalid = 0

for d in data:
    try:
        float(d)
        valid += 1
    except:
        invalid +=1

print('valid:', valid)
print('invalid:', invalid)
```

```python
#Sum Only Valid Numbers
data = ["10", "20", "abc", "30"]

total = 0

for d in data:
    try:
        total += float(d)
    except:
        pass

print(total)
```

```python
#Safe Average (Avoid Division by Zero)
data = []

total = sum(data)
count = len(data)

average = total / count if count > 0 else 0

print(average)
```

## PART 11 REAL CASE CONDITIONS

```python
#Age + ID Validation
ages = [17, 20, 25, 16, 30]
has_id = [True, True, False, True, True]

for i in range(len(ages)):
    if ages[i] >= 18 and has_id[i] == True:
        print(f'Person {i} allowed')
```

```python
#Member or Coupon
is_member = [True, False, True, False, True]
has_coupon = [False, True, False, False, True]

for i in range(len(is_member)):
    if is_member[i] == True or has_coupon[i] == True:
        print(f'User {i} gets discount')
```

```python
#Discount (zip + enumerate)

is_member = [True, False, True, False, True]
has_coupon = [False, True, False, False, True]

for i, (m,c) in enumerate(zip(is_member, has_coupon)):
    if m or c:
        print(f'user {i} gets discount')
```

```python
#Temperature Filter

temperatures = [30, 32, 35, 40, 50, 100, 20]

for t in temperatures:
    if t >= 50:
        print(t)
```

```python
#Temperature (while)

temperatures = [30, 32, 35, 40, 50, 100, 20]

i = 0

while i < len(temperatures):
    if temperatures[i] >= 50:
        print(temperatures[i])
    i += 1
```
