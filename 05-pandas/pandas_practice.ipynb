```python
import numpy as np

df2.loc[2, 'Marks'] = np.nan
df2.loc[5, 'Age'] = np.nan
df2.loc[10, 'Marks'] = np.nan

print(df2.info())
```

```python
print(df2)
```

```python
mean_marks = df2['Marks'].mean()
print(f'Rata-rata Marks: {mean_marks}')

#isi Nan dengan rata-rata
df2['Marks'].fillna(mean_marks, inplace=True)
print(df2)
```

```python
df2['Marks'] = df2['Marks'].fillna(mean_marks)
print(df2)
```

```python
mean_age = df2['Age'].mean()
df2['Age'] = df2['Age'].fillna(mean_age)
print(df2)
```

```python
df.loc[3, 'Marks'] = np.nan

print('before:')
print(df2[df2['Marks'].isna()])

df_clean = df2.dropna()

print('Sesudah:')
print(df_clean.info())
```
