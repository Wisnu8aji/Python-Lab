```python
data = {
    'name': ['Budi', 'Sari', 'Andi', 'Dewi', 'Riko'],
    'salary': [8000000, 9500000, 7500000, 11000000, 8750000],
    'department': ['Data', 'AI', 'Data', 'AI', 'Engineering']
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df['name'])
```

```python
print(df[['name']])
```

```python
print(df[['name', 'salary']])
```

```python
print(df.iloc[2])
```

```python
print(df.loc[2])
```

```python
df2 = df.set_index('name')
print(df2)
```

```python
print(df2.iloc[2])
print(df2.loc['Andi'])
```

```python
print(df2.loc['Dewi', 'salary'])
```

```python
print(df.iloc[0:3])
```

```python
print(df.loc[0:3])
```

```python
print(df[df['salary'] > 9000000])
```

```python
for i in df['salary']:
    if i > 9000000:
        print(i)
```

```python
print(df['department'].unique())
```

```python
print(df['department'].nunique())
```

```python
high_salary = df[df['salary']>9000000]
high_salary.to_csv('high_salary_report.csv', index=False)
print('saved!')
```
