```python
c = np.array([20,1,2,3,4])
print('sebelum:', c)

c[0] = 100
print('sesudah ubah index 0:', c)

c[4] = 100
print('sesudah ubah index 4:', c)
```

```python
arr = np.array([1,2,3,4,5,6,7])

d = arr[1:4]
print('slice 1:4 :', d)

print('slice : 4 =', arr[:4])

print('slice 4: =', arr[4:])

print('slice ::2 =', arr[::2])
```

```python
salary = np.array([5,8,12,7,15, 9, 6, 11, 14, 10])

print('mean = ', salary.mean())

print('std =', salary.std())

print('max =', salary.max())

print('min =', salary.min())
```

```python
u = np.array([1, 2, 3])
v = np.array([4,5,6])


print('addition =', u + v)
print('substraction =', u-v)
print('multiply =', u * v)
print('division =', np.divide(u, v))
```
