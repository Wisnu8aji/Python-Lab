```python
import matplotlib.pyplot as plt

#generate 100 points from 0 to 2pi
x = np.linspace(0, 2*np.pi, num=100)

# Apply sine function
y = np.sin(x)

#plot the graph
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x (radians)')
plt.ylabel('sin(x)')
plt.show()
```

```python
error_rate = np.array([0.85, 0.76, 0.91, 0.88, 0.72, 0.95, 0.80, 0.83])

print('Mean =',error_rate.mean())
print('std deviation =', error_rate.std())
print('Max =',error_rate.max())
print('Min =',error_rate.min())
print('high error days =', error_rate[error_rate > 0.85])
```
